# LinearApproximations
In this section, we explore the idea of approximating potentially complicated functions with a much simpler family of functions: linear functions.

Let's first remind ourselves of the following version of the definition of the derivative at a point: $$f'(a)=\displaystyle\lim_{x\to a}\frac{f(x)-f(a)}{x-a}.$$

In defining the derivative with a limit, what we are saying is that for $x$-values \textit{very close} to $a$ we have: $$f'(a)\approx\frac{f(x)-f(a)}{x-a}.$$

To be clear, we are saying we can \textit{approximate} the value of the derivative at a point with the above expression for $x$-values \textit{very close} to $a$.

We can then rearrange the above approximation to read $$f(x)\approx f'(a)(x-a)+f(a)$$ for $x$-values very close to $a$.

Now wait just a minute - this looks strikingly familiar. This is nothing but the equation of the tangent line to $f(x)$ at $x=a$ so let's give it a special name.

Let $f(x)$ be differentiable at $x=a$. Then the linear approximation $L_a^f(x)$, to $f(x)$ at $x=a$ is the function $$L_a^f(x)=f'(a)(x-a)+f(a)$$ where if $f(x)$ is clear from context, we write instead $L_a(x)$.

Let $f(x)=\sqrt{x}$. Use an appropriate linearization to approximate $\sqrt{3.98}$ and $\sqrt{4.05}$.
We'll begin by looking at the actual (so called calculator values) of $\sqrt{3.98}$ and $\sqrt{4.05}$.

```python
import math

x1 = 3.98
x2 = 4.05

sqrt1 = math.sqrt(x1)
sqrt2 = math.sqrt(x2)

print(f"√{x1} = {sqrt1}")
print(f"√{x2} = {sqrt2}")
```

Next, we will use an appropriate linearization to approximate $\sqrt{3.98}$ and $\sqrt{4.05}$. Since both values in the squareroot is close to $4$, we choose $a=4$. 

```python
f = lambda x: x**0.5
f_prime = lambda x: 1 / (2 * x**0.5)

a = 4
x1 = 3.98
x2 = 4.05

L = lambda x: f(a) + f_prime(a) * (x - a)

print(f"Linear approximation of √{x1} ≈ {L(x1)}")
print(f"Linear approximation of √{x2} ≈ {L(x2)}")
```

Now, let's compare our calculations:

```python
# Define the function and its derivative
f = lambda x: x**0.5
f_prime = lambda x: 1 / (2 * x**0.5)

a = 4  # point of linearization

# Linear approximation function
L = lambda x: f(a) + f_prime(a) * (x - a)

# Points to evaluate
x_values = [3.98, 4.05]

print("x\tLinear Approx\tActual sqrt\tDifference")
for x in x_values:
    linear_approx = L(x)
    actual = f(x)
    diff = abs(actual - linear_approx)
    print(f"{x}\t{linear_approx:.6f}\t{actual:.6f}\t{diff:.6f}")
```

Overall, our approximations were pretty good! We can see we \textit{overestimated} the true values of both $\sqrt{3.98}$ and $\sqrt{4.05}$, which we can see graphically by noting that the tangent line at $x=4$ lies above $f(x)$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define function and derivative using exponent 0.5
f = lambda x: x**0.5
f_prime = lambda x: 1 / (2 * x**0.5)

a = 4  # point of linearization

# Linear approximation function
L = lambda x: f(a) + f_prime(a) * (x - a)

# Create x values around 4 (for example, from 3.5 to 4.5)
x_vals = np.linspace(2, 6, 500)

# Compute f(x) and L(x)
f_vals = f(x_vals)
L_vals = L(x_vals)

# Plot the function and its linear approximation
plt.figure(figsize=(8,5))
plt.plot(x_vals, f_vals, label=r'$f(x) = \sqrt{x}$', color='blue')
plt.plot(x_vals, L_vals, label=r'Tangent line at $x=4$', color='red', linestyle='--')

# Mark the point of tangency
plt.scatter([a], [f(a)], color='black')
plt.text(a, f(a)+0.05, r'$(4, 2)$', fontsize=12)

# Highlight points of interest
points_x = [3.98, 4.05]
points_f = [f(x) for x in points_x]
points_L = [L(x) for x in points_x]

plt.scatter(points_x, points_f, color='blue', marker='o', label=r'Actual $\sqrt{x}$ points')
plt.scatter(points_x, points_L, color='red', marker='x', label=r'Linear approx points')

# Labels and title
plt.title(r'Square Root Function and Its Linear Approximation at $x=4$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
